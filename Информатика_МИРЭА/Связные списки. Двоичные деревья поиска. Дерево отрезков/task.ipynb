```python
class Node: # определяем класс и указываем атрибуты
    def __init__(self, data):
        self.data = data
        self.next = None 
        self.prev = None


class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None
    

    def append(self, data):
        new_node = Node(data)
        if not self.head:  # Если список пуст
            self.head = new_node
            new_node.next = new_node  # Замыкаем на себя
            new_node.prev = new_node  # Замыкаем на себя
        else:
            tail = self.head.prev  # Текущий хвостовой узел
            tail.next = new_node  # Хвост указывает на новый узел
            new_node.prev = tail  # Новый узел указывает на хвост
            new_node.next = self.head  # Новый узел указывает на голову
            self.head.prev = new_node  # Голова указывает на новый узел


    # Удаление узла с заданным значением
    def delete(self, data):
        if not self.head:
            return

        current = self.head

        while True:
            if current.data == data:
                # Если список из одного элемента
                if current.next == current:
                    self.head = None
                    break

                # Удаление головы
                if current == self.head:
                    self.head = current.next

                # Обновление указателей соседних узлов
                current.prev.next = current.next
                current.next.prev = current.prev
                break

            current = current.next
            if current == self.head:
                break

    # Обход списка и вывод элементов
    def display(self):
        if not self.head:
            print("Список пуст")
            return

        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()


cll = CircularDoublyLinkedList()
cll.append(1)
cll.append(2)
cll.append(3)
cll.append(4)

print("Исходный список:")
cll.display()

cll.delete(2)
print("После удаления 2:")
cll.display()

cll.delete(1)
print("После удаления 1:")
cll.display()
```
